```python
def load_df(direction, file):
    data = pd.read_csv(f"{direction}{file}")
    return data
```

```python
data = load_df(direction, file)
```

```python
def plot_pie_chart(column, data):
    #pie_db = data.groupby(by=column).agg({"amount": ["sum", "count"]})
    pie_db = data.groupby(by=column).sum()
    column_list = [ x for x in pie_db.columns.tolist() if x !="amount"]
    pie_db = pie_db.drop(columns=column_list)    
    pie_db["count"] = data.groupby(by=column).count()["amount"]
    plot = pie_db.plot.pie(y="amount", figsize=(5, 5))
    print(pie_db)
    return
```

```python
def evaluate_organics(data):
    organics = data.dropna()
    organic_percentage = round(np.mean(np.mean(data["organic percentage"])),2)
    print(f'The percentage of organic \ngrocery products is {np.mean(data["organic percentage"])}')
    return organic_percentage
```

```python
def eval_month(data, month, year):
    year_df = data[data["year"] == year]
    month_df = year_df[data["month"] == month]
    summed_expenses = round(np.sum(month_df["amount"]),2)
    plot_pie_chart("category", month_df)
    print(f"------------------------- \n I total {summed_expenses} € have \nbeen spent this month.\n-------------------------")
    evaluate_organics(month_df)
    return summed_expenses
```

```python
eval_month(data, 1, 2021)
```
